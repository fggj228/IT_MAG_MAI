```python
def TenToTwo(n): 
    b = ''
    n = int(n)
    while n > 0:
        b =str(n % 2) + b
        n = n // 2
        
    value = list(b)
    while len(value) < 8:
        value.insert(0,'0')
    return value


dop = TenToTwo(input_number)
print (dop)
```
