```python
from time import sleep
from random import randint
from IPython.display import clear_output #Только для jupyter notebook

k = 10
world = []
 
for line in range(k):
    line = []
    for index in range(k):
        cell = [randint(0, 1), 0]
        line.append(cell)
    world.append(line)
 
 
def print_lines():          # функция, выводящая состояние поля в удобном виде
    for i in range(k):
        for j in range(k):
            print(world[i][j][0], end=" ")
        print()
    print()
 
 
def counter():
    for i in range(k):             
        for j in range(k):       # перебираем все клетки
 
            x_from, y_from = i - 1, j - 1       
            x_to, y_to = i + 1, j + 1                
 
            if x_from < 0:    
                x_from = 0
 
            if y_from < 0:
                y_from = 0
 
            if x_to > k - 1:
                x_to = k - 1
 
            if y_to > k - 1:
                y_to = k - 1      
 
            summ = 0
            for x in range(x_from, x_to + 1): 
                for y in range(y_from, y_to + 1):       # перебираем квадрат 3*3 и считаем сколько там "живых" клеток 
                    summ += world[x][y][0]               
            summ -= world[i][j][0]
            world[i][j][1] = summ
 
 
def step():            # пробегаемся по массиву и в зависимости от кол-ва соседей клетки меняем (или не меняем) ее состояние 
    for i in range(k):
        for j in range(k):
            if world[i][j][1] > 3 or world[i][j][1] < 2:
                world[i][j][0] = 0
            else:
                world[i][j][0] = 1
    print_lines()
    counter()
 
 
def life_counter():        # пробегаемся по массиву и суммируем живые клетки
    c = 0
    for i in range(k):
        for j in range(k):
            c += world[i][j][0]
    return c

 
#print_lines()




counter()

while life_counter() > 0:
    step()
    sleep(2.5)
    clear_output() #Для очистки экрана
```
